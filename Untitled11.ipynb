```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np  # Import numpy

# Alpha Vantage API setup
API_KEY = '6BAS9SYLIURHE464'
BASE_URL = 'https://www.alphavantage.co/query'

# Function to get stock data
def get_stock_data(symbol, interval='daily'):
    params = {
        'function': 'TIME_SERIES_DAILY',
        'symbol': symbol,
        'apikey': API_KEY
    }
    response = requests.get(BASE_URL, params=params)
    data = response.json()
    return data

# Parse the data into a DataFrame
def parse_data(data):
    time_series = data['Time Series (Daily)']
    df = pd.DataFrame.from_dict(time_series, orient='index')
    df = df.astype(float)
    df.index = pd.to_datetime(df.index)
    df.sort_index(inplace=True)
    return df

# Function to perform basic analysis
def analyze_stock(symbol):
    data = get_stock_data(symbol)
    df = parse_data(data)

    # Calculate moving averages
    df['SMA_50'] = df['4. close'].rolling(window=50).mean()
    df['SMA_200'] = df['4. close'].rolling(window=200).mean()

    # Plot the data
    plt.figure(figsize=(14, 7))
    plt.plot(df['4. close'], label='Close Price')
    plt.plot(df['SMA_50'], label='50-day SMA')
    plt.plot(df['SMA_200'], label='200-day SMA')
    plt.title(f'Stock Analysis for {symbol}')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.show()

    # Basic strategy: Golden Cross
    df['Signal'] = 0
    signal_values = np.where(df['SMA_50'][50:] > df['SMA_200'][50:], 1, 0)
    df.loc[df.index[50:], 'Signal'] = signal_values
    df['Position'] = df['Signal'].diff()

    print(df[['4. close', 'SMA_50', 'SMA_200', 'Signal', 'Position']].tail())

# Example usage
analyze_stock('TSLA')
```
